```python
import gradio as gr
import joblib
import pandas as pd
import datetime

model = joblib.load("rf_beras_surabaya_2024.joblib")

def predict_price(input_date, price_lag1):

    dt = pd.to_datetime(input_date)

    day_number = (dt - pd.Timestamp(2024,1,1)).days
    day_of_week = dt.dayofweek
    month = dt.month
    
    df = pd.DataFrame({
        "day_number": [day_number],
        "day_of_week": [day_of_week],
        "month": [month],
        "price_lag1": [price_lag1]
    })
    
    pred = model.predict(df)[0]
    return float(pred)

app = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Textbox(label="Tanggal (YYYY-MM-DD)"),
        gr.Number(label="Harga Kemarin (price_lag1)")
    ],
    outputs=gr.Number(label="Prediksi Harga (Rp)")
)

app.launch(share=True)
```
